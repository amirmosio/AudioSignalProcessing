## Part 1

IDFT is as follows:


$x[n]=\sum_{i=0}^{N-1}{ X[k]  e^{j\frac{2\pi}{N}kn}}$


## Part 2

```python
import math
import numpy as np

def fft(vector):
  N = len(vector)
  res = np.zeros(N)
  for k in range(N):
    res[k] = np.sum(vector * np.exp(-2j * np.pi * k * np.arange(N)/N))
  return res

print(fft(np.exp(2j * np.pi * np.arange(8) / 8)))
```

## Part 3

```python
import math
import numpy as np

def ifft(vector):
  N = len(vector)
  res = np.zeros(N)
  for n in range(N):
    res[n] = np.sum(vector * np.exp(2j * np.pi * n * np.arange(N)/N)) / N
  return res

print(ifft(np.array([10, -2, -2, -2])))
```

## Part 8

With constant N, increasing T means decreasing sampling frequency results in distnat $\sigma$ pulse until **Nyquist** frequency. after that we have information loss.

With T constant, increasing N results in ideal pulse signals and sharp edges.

```python
T_range = [1/10, 1/5, 1/2, 1, 2, 4]
N_range = [50, 100, 200, 400, 600, 1000, 2000, 3000]
fig, axs = plt.subplots(len(N_range), len(T_range), figsize=(30,30))
plt_number = 0

for T_i in range(len(T_range)):
  for N_i in range(len(N_range)):
    T = T_range[T_i]
    N = N_range[N_i]
    n_range = np.arange(-1 * N//2, N//2+1)
    x_vector = np.array([x(T*n) for n in n_range])
    x_fft = fft(x_vector).astype(np.float)

    axs[N_i][T_i].plot(x_fft)
    axs[N_i][T_i].title.set_text(f'T={T}, N={N}')
    plt_number +=1
```
