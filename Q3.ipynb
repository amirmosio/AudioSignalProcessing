CONVERTING TO FREQUENCY DOMAIN AND CALCULATING FREQUENCY RESPOSNE

$ DistortedInFrequencyDomain = FrequencyRespose.CleanInFrequencyDomain$ 

```python
clean1_fft = scipy.fft.fft(clean1_signal)
clean1_freq = scipy.fft.fftfreq(len(clean1_fft), (1.0/sample_rate)) 

distorted1_fft = scipy.fft.fft(distorted1_signal)
distorted1_freq = scipy.fft.fftfreq(len(distorted1_fft), (1.0/sample_rate)) 


frequency_response = distorted1_fft/clean1_fft
frequency_response_freq = scipy.fft.fftfreq(len(frequency_response), (1.0/sample_rate)) 
```
